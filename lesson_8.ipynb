### Задание 1

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
boston = load_boston()

data = boston.data
feature_names = boston.feature_names
```

```python
x = pd.DataFrame(data, columns=feature_names)
```

```python
x.head()
```

```python
price = boston.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, price, test_size=0.2, random_state=42)
```

```python
x_train.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_sc = scaler.fit_transform(x_train)
x_test_sc = scaler.transform(x_test)
```

```python
x_train_sc = pd.DataFrame(x_train_sc, columns=feature_names)
x_test_sc = pd.DataFrame(x_test_sc, columns=feature_names)
```

```python
x_train_sc.describe()
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, learning_rate=250, random_state=42)
```

```python
x_train_tsn = tsne.fit_transform(x_train_sc) 
```

```python
x_train_tsn
```

```python
plt.scatter(x_train_tsn[:,0], x_train_tsn[:,1]);
```

### Задание 2

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3, max_iter=100, random_state=42)
```

```python
train_model = model.fit_predict(x_train_tsn)
```
